```python
#Searching for clients with name starting with "Vet"

search_query2='Vet'
url2 = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, search_query2, RADIUS, LIMIT)
url2

results2 = requests.get(url2).json()
results2

# assign relevant part of JSON to venues
venues2 = results2['response']['venues']

# tranform venues into a dataframe
dataframe2 = json_normalize(venues2)
dataframe2.head()
```

```python
dataframe2.shape
```

```python
#Looking for clients with name including "Perro" (spanish for dog)

search_query3='Perro'
url3 = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, search_query3, RADIUS, LIMIT)
url3

results3 = requests.get(url3).json()
results3

# assign relevant part of JSON to venues
venues3 = results3['response']['venues']

# tranform venues into a dataframe
dataframe3 = json_normalize(venues3)
dataframe3.head()
```

```python
#Looking for clients with name including "Gato" (spanish for cat)

search_query4='Gato'
url4 = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, latitude, longitude, VERSION, search_query4, RADIUS, LIMIT)
url4

results4 = requests.get(url4).json()
results4

# assign relevant part of JSON to venues
venues4 = results4['response']['venues']

# tranform venues into a dataframe
dataframe4 = json_normalize(venues4)
dataframe4.head()
```

```python
#Concatenating results
frames = [dataframe, dataframe2, dataframe3, dataframe4]
clientes=pd.concat(frames)
clientes.shape
```

```python
# keep only necessary columns
filtered_columns = ['name','location.distance', 'location.lat','location.lng', 'location.formattedAddress' ]  + ['id']
clientes_resumen = clientes.loc[:, filtered_columns]
clientes_resumen.head()
```

```python
#Preparing data for cluster
clientes_cluster= clientes_resumen.drop(['name','location.formattedAddress','id', 'location.distance'], 1)
clientes_cluster.head()
```

```python
#Clustering in 3 clusters

# set number of clusters
kclusters = 3

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(clientes_cluster)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

```python
# add clustering labels
clientes_resumen.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
clientes_resumen.head()
```

```python
#Visualization
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(clientes_resumen['location.lat'], clientes_resumen['location.lng'], clientes_resumen['name'], clientes_resumen['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
